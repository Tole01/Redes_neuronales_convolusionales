# Funcionamiento del Kernell

La funcion `nd.convolve()` nos ayuda a ver como es que funciona el kernel. Para esto, debes usar por parametro una imagen convertida en escala de grises y la matriz de kernell para detectar los bordes.

```python
import scipy.ndimage as nd 
from skimage import  io, color 

# Definimos la matriz de la imagen en escala de grises
img_gray = color.rgb2gray(matriz_mara)

# Creamos un kernell para detectar bordes verticales
kernel = np.array([[1,0,-1],
                  [1,0,-1],
                  [1,0,-1]])

# Usamos la funcion convolucion para observar el funcionamiento del kernell
img_new = nd.convolve(img_gray, kernel)

# Graficamos la diferencia
fig, axes = plt.subplots(1,2, figsize = (15,10))
axes[0].imshow(img_gray, cmap=plt.cm.gray)
axes[0].set_title('Original')
axes[0].axis('off')

axes[1].imshow(img_new, cmap=plt.cm.gray)
axes[1].set_title('Convolution')
axes[1].axis('off')
```

# Data Augmentation

Esta librerias importa los modulos necesarios para hacer data augmentation a partir del modulo ImageDataGenerator. El resto de los modulos permiten guardar las imagenes generadas por este proceso.


```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
# Creamos el modulo ImageDataGenerator con los diferentes parametros
datagen = ImageDataGenerator(rotation_range= 40,
                  width_shift_range= 0.2 ,
                  height_shift_range= 0.2,
                  zoom_range= 0.2,
                  horizontal_flip= True,
                  fill_mode = 'nearest',
                  brightness_range= [0.4, 1.5])

# Cambia la imagen a float y hacemos reshape
x = img_to_array(matriz_mara)
print(x.shape)
x = x.reshape((1,) + x.shape)
x.shape
```

```python
i = 0

# Creamos iterador para 10 imagenes diferentes
for batch in datagen.flow(x, batch_size = 1):
    plt.figure(i)
```
